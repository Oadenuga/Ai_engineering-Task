```python

for col in just_number.columns:
    min_value = df2[col].min()
    max_value = df2[col].max()
    mean = df2[col].mean()
    median = df2[col].median()
    mode = df2[col].mode().iloc[0]
    value_range = max_value - min_value
    print(f"For {col}")
    print(f"             Maximum value: {max_value}")
    print(f"             Minimum value: {min_value}")
    print(f"             Mean: {mean:.1f}")
    print(f"             Median: {median:.1f}")
    print(f"             Mode: {mode}")
    print(f"             Range: {value_range}")
    print()
```

#### General Interpretation

1)Age

Range: 1–9 years, mean = 3.8 years.

The distribution is skewed toward younger ages since the median (3.0) and mode (3.0) are lower than the mean (3.8).

Suggests most individuals are young, but a few older ones raise the mean.

2)Head Length (hdlngth)

Range: 82.5–103.1 mm, mean ≈ 92.6 mm.

Median and mode are close to the mean, indicating a fairly symmetric distribution.

Variation is moderate (range ≈ 20.6 mm).

3)Skull Width (skullw)

Range: 50.0–68.6 mm, mean = 56.9 mm.

Median (56.3) and mode (57.6) are close to the mean → balanced distribution.

Indicates consistent skull proportions among individuals.

4)Total Length (totlngth)

Range: 75–96.5 cm, mean = 87.1 cm.

Median (88.0) and mode (89.0) slightly higher than mean → slight left skew.

Shows a fairly uniform body size.

5)Tail Length (taill)

Range: 32–43 cm, mean = 37 cm.

Mean, median, and mode are nearly the same → symmetric distribution.

Suggests very consistent tail sizes.

6)Foot Length (footlgth)

Range: 60.3–77.9 mm, mean = 68.5 mm.

Median (68.0) and mode (73.2) are slightly different, suggesting mild variation.

Range (≈ 17.6 mm) shows some diversity in foot size.

7)Ear Conch (earconch)

Range: 40.3–56.2 mm, mean = 48.1 mm.

Median (46.8) < mean, mode (44.9) < mean → right-skewed distribution.

A few individuals with larger ear sizes increase the mean.

8)Eye Diameter (eye)

Range: 12.8–17.8 mm, mean = 15.0 mm.

Median (14.9) and mode (14.4) close → balanced.

Very small range (5 mm) → highly consistent eye sizes.

9)Chest

Range: 22–32 cm, mean = 27 cm.

Mean, median, and mode overlap → symmetrical.

Indicates a stable chest size across individuals.

10)Belly

Range: 25–40 cm, mean ≈ 32.6 cm.

Median (32.5) and mode (32.0) align with mean → symmetry.

Some variation, but overall balanced body shape.

#### Key Observations

Consistency: Traits like tail length, chest, eye size, and belly show strong symmetry (mean ≈ median ≈ mode). This suggests these measurements are stable and less influenced by outliers.

Variation: Age, foot length, and ear conch show noticeable differences between mean, median, and mode, suggesting skewness or more variation in those traits.

Biological implication (if this is for animals):

Younger individuals dominate the sample (most are around 3 years).

Core body proportions (head, skull, chest, belly, tail) are stable, which could indicate developmental uniformity.

Peripheral traits (ear, foot) show more variation, possibly due to environmental or genetic factors.
