```python

from pydantic import BaseModel, Field, ValidationError,EmailStr, field_validator
from typing import Optional
from datetime import datetime


class user_name(BaseModel):
    user_user: str= Field(..., min_length=3, max_length=15,pattern=r"^[A-Za-z0-9_]+$")
    email: EmailStr
    password: str=Field(..., min_length=6, max_length= 15, pattern=r"^[A-Za-z0-9_]+$")
    post: str=Field(..., min_length=1, max_length=280, description="post content(1-280 characters)")
    created_at: datetime = Field(default_factory=datetime.now)
    
    
```

```python
def twitter_demo():
    print("Username Validation")
    try:
        user_name(user_user= "peluxstar", email="adenuga.ap@gmail.com", password="pelumi4life", post= "my firrst validation")
    except ValidationError as e:
        print(e)
```
